The aim of this notebook is to perform exploratory analysis on the data.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# reading in data
data = pd.read_csv("candy-data.csv")
```

```python
# get shape of dataframe plus file size
data.info()
```

```python
# checking null values per column
data.isnull().sum()
```

```python
# check summary statistics of columns
print(data.describe())
```

```python
non_binary_cols = data.select_dtypes(include="float64").columns.tolist()
non_binary_cols
```

```python
# visualize the relationship between the target and non-binary features
non_binary_cols = data.select_dtypes(include="float64").columns.tolist()
sns.pairplot(data, y_vars = non_binary_cols)
plt.show()
```

```python
# plotting histogram of winpercent for presentation
sns.displot(data["winpercent"], kind="hist", color="#FFF000")
plt.title("The distribution of winpercent is right-skewed.")
plt.xlabel("Win percent (% of matchups)")
plt.show()
```
